```python
import pandas as pd
import folium
```

```python
df= pd.read_csv("서울시기관_kakaomap.csv", encoding='CP949')
```

```python
df
```

```python
df.info()
```

```python
lat= df['Latitude'].mean()
long = df['Longitude'].mean()
#지도의 중심을 지정하기 위해 위도의 경도의 평균을 구한다
lat, long
```

```python
m= folium.Map([lat, long], zoom_start=11)
```

```python
m
```

```python
df['Location_Dep'].values
```

```python
df.loc[df['Location_Dep']=='서울시 본청', 'Latitude']
```

```python
for i in df['Location_Dep'].values:
    sub_lat= df.loc[df['Location_Dep']==i,'Latitude']
    sub_long= df.loc[df['Location_Dep']==i,'Longitude']
    title= f"{i} - {df.loc[df['Location_Dep']==i,'Location_Regex'].values[0]}"
    icon_color='blue'
        
    folium.Marker(
        [sub_lat, sub_long], 
        icon= folium.Icon(color=icon_color),
        popup=f'<i>{title}</i>', tooltip=title).add_to(m)

m
```

```python
m.save('Seoul_Map.html')
```

```python
from folium.plugins import MarkerCluster

m2= folium.Map([lat, long], zoom_start=8, tiles='stamen toner')
marker_cluster= MarkerCluster().add_to(m2)

for i in df['Location_Dep'].values:
    sub_lat= df.loc[df['Location_Dep']==i,'Latitude']
    sub_long= df.loc[df['Location_Dep']==i,'Longitude']
    title= f"{i} - {df.loc[df['Location_Dep']==i,'Location_Regex'].values[0]}"
    icon_color='blue'

    
    folium.Marker(
        [sub_lat, sub_long], 
        icon= folium.Icon(color=icon_color),
        popup=f'<i>{title}</i>', tooltip=title).add_to(marker_cluster)


m2
```

```python
m2.save('Seoul_Map_MarkerCluster.html')
```
