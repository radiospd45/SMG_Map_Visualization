```python
Location_total_2=[]
Non_total=[]
for i in Location_total:
    if p_.match(i) == None:
        non_group= i
        Non_total.append(non_group)
    else:
        ma_group= p_.match(i).group()
        Location_total_2.append(ma_group)
```

* 괄호제거는 나중에 해주자

```python
def get_location(string):
    if string.startswith('서울특별시'):
        result= p_.match(string).group()
        return result
    elif string.startswith('경기도'):
        result= k.match(string).group()
        return result
    elif string.startswith('강원도'):
        result= s.match(string).group()
        return result
```

```python
df['Location_Regex'] = df['Location_Re'].apply(lambda x:x if p_.match(x) == None else get_location(x))
```
